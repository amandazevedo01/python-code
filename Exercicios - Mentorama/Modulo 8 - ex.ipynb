OBJETO	DESCRIÇÃO	comandos
Thread	Objeto que representa um único thread de execução.	threading.active_count()¶
Lock	Objeto de bloqueio primitivo.	Lock.acquire()
RLock	O objeto de bloqueio RLock ou Reentrante fornece capacidade para um único thread (re) adquirir um bloqueio já mantido (bloqueio recursivo).	RLock.acquire()
Condition	Objeto de variável de condição faz com que um thread espere até que certa "condição" seja satisfeita por outro thread (como mudança de estado ou algum valor de dados)	Condition.wait()
Event	É uma versão mais geral das variáveis de condição, em que várias threads podem ser feitas para aguardar a ocorrência de algum evento e todas as threads em espera só serão ativadas quando o evento acontecer.	threading.Event
Semaphore	Fornece um "contador" de recursos finitos compartilhados entre blocos de threads quando nenhum está disponível.	threading.Semaphore(value=1)
BoundedSemaphore	Semelhante a um semáforo, mas garante que nunca exceda seu valor inicial.	threading.BoundedSemaphore(value=1)¶
Timer	Semelhante ao Thread, exceto que ele espera por um período de tempo especificado antes	Timer(30.0, hello)
	de ser executado.	
Barrier	Cria uma "barreira" na qual um número especificado de threads deve chegar antes que todos	Barrier(2, timeout=5)
	tenham permissão para continuar.
